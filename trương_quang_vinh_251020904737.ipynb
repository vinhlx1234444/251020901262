```python
s = input("Nhập vào một chuỗi: ")
```

```python
tu_list = s.split()
```

```python
so_tu = len(tu_list)
```

```python
tu_dai_nhat = tu_list[0]
```

```python
for tu in tu_list:
    if len(tu) > len(tu_dai_nhat):
        tu_dai_nhat = tu
```

```python
print("Số từ trong chuỗi:", so_tu)
print("Từ dài nhất:", tu_dai_nhat)
```

```python
input_str = input("Nhập danh sách điểm: ")
```

```python
diem_list = [float(x) for x in input_str.split()]
```

```python
diem_hop_le = []
for diem in diem_list:
    if 0 <= diem <= 10:
        diem_hop_le.append(diem)
```

```python
if len(diem_hop_le) == 0:
    print("Không có điểm hợp lệ")
else:
    trung_binh = sum(diem_hop_le) / len(diem_hop_le)
```

```python
so_dat = 0
for diem in diem_hop_le:
    if diem >= 5:
        so_dat += 1
```

```python
print("Average:", trung_binh)
print("Passed:", so_dat)
```
